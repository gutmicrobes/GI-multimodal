```python
storage_path = r'F:\CRC&STAD_image\image_class_sample'
access_path1 = r'F:\CRC&STAD_image\MSIMUT'
access_path2 = r'F:\CRC&STAD_image\MSS'
```

```python
num = []  #num是把所有在两个文件夹里拥有图片的样本id和图片数量保存的list
finaly_sample = []   #所有在两个文件夹里拥有图片的样本id
count = 0  
for key in sample_name:
    count+=1
    print(count)
    n = 0
    key1 = key[0:4]+'-'+key[5:7]+'-'+key[8:12]
    #os.mkdir(storage_path+os.sep+key)
    for file in os.listdir(access_path1):
        if key1 in file:
            n+=1
            finaly_sample.append(key)
            break
            #shutil.copy(access_path1+os.sep+file,storage_path+os.sep+key+os.sep+file)
    for file in os.listdir(access_path2):
        if key1 in file:
            n+=1
            finaly_sample.append(key)
            break
            #shutil.copy(access_path2+os.sep+file,storage_path+os.sep+key+os.sep+file)
    if n!=0:
        num+=[[key,n]]
```
