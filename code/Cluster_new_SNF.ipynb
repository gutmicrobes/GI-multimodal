```python
import os
import pandas as pd
import numpy as np
import math
```

```python
def power_mat(data):
    mean_dis = []  #定义样本与其他样本的平均距离的列表
    n = data.shape[0]
    for i in range(n):
        sum1 = 0
        for j in range(n):
            if i!=j:
                sum1+=data.iloc[i,j]
        mean_dis.append(sum1/(n-1))
    p = pd.DataFrame(np.zeros([n,n]))
    u = 0.5
    for i in range(n):
        for j in range(n):
            v = (mean_dis[i]+mean_dis[j]+data.iloc[i,j])/3
            p.iloc[i,j] = math.exp(-pow(data.iloc[i,j],2)/(u*v))
    p0 = pd.DataFrame(np.zeros([n,n]))
    for i in range(n):
        sum2 = sum(p.iloc[i,:])
        for j in range(n):
            if i == j:
                p0.iloc[i,j] = 0.5
            else:
                p0.iloc[i,j] = p.iloc[i,j]/(2*(sum2-p.iloc[i,i]))
    return p,p0
#p0是权重矩阵，p是高斯核矩阵
```

```python
from sklearn.cluster import KMeans
from sklearn.metrics import silhouette_score
import matplotlib.pyplot as plt
def pre_cluster(data,num):
    silhouette_list=[]
    sse_list=[]
    title = ['mRNA','miRNA','Image','Cluster']
    for i in range(2,10):
        kmeans=KMeans(n_clusters=i,init='k-means++',random_state=0).fit(data)
        cluster_labels=kmeans.predict(data)
        #print(len(cluster_labels))
        silhouette_score_avg=silhouette_score(data,cluster_labels) #轮廓系数
        sse=kmeans.inertia_  #簇内平方和
        silhouette_list.append(silhouette_score_avg)
        sse_list.append(sse)
    plt.plot(range(2,10),sse_list,color='red',marker='o',markersize=5,label='sse')
    plt.ylabel("SSE")
    plt.xlabel("K")
    plt.title(title[num])
    plt.show()
    k = 0
    m = 0
    #print(sse_list)
    for i in range(1,5):
        #print((sse_list[i-1]-sse_list[i])/(sse_list[i]-sse_list[i+1]))
        if (sse_list[i-1]-sse_list[i])/(sse_list[i]-sse_list[i+1])>m:
            m = (sse_list[i-1]-sse_list[i])/(sse_list[i]-sse_list[i+1])
            k = i+2
    kmeans=KMeans(n_clusters=k,init='k-means++',random_state=0).fit(data)
    cluster_labels=kmeans.predict(data)
    silhouette_score_avg=silhouette_score(data,cluster_labels) #轮廓系数
    sse=kmeans.inertia_
    #print(cluster_labels)
    return k,sse,silhouette_score_avg,cluster_labels,sse_list
```

```python
def core_mat(data,data_list):
    n = data.shape[0]
    S = pd.DataFrame(np.zeros([n,n]))
    for i in range(n):
        sum2 = 0
        for j in range(n):
            if data_list[i] == data_list[j]:
                sum2+=data.iloc[i,j]
        for j in range(n):
            if data_list[i] == data_list[j]:
                S.iloc[i,j] = data.iloc[i,j]/sum2

    return S
                
            
```

```python
def max_(loss):
    n = len(loss)
    k = 0
    m = 0
    for i in range(1,n-1):
        if (loss[i-1]-loss[i])/(loss[i]-loss[i+1])>m:
            m = (loss[i-1]-loss[i])/(loss[i]-loss[i+1])
            k  = i+1
    return k
```

```python
#谱聚类
def calLaplacianMatrix(adjacentMatrix):

    # compute the Degree Matrix: D=sum(A)
    degreeMatrix = np.sum(adjacentMatrix, axis=1)

    # compute the Laplacian Matrix: L=D-A
    laplacianMatrix = np.diag(degreeMatrix) - adjacentMatrix

    # normailze
    # D^(-1/2) L D^(-1/2)
    sqrtDegreeMatrix = np.diag(1.0 / (degreeMatrix ** (0.5)))
    return np.dot(np.dot(sqrtDegreeMatrix, laplacianMatrix), sqrtDegreeMatrix)
```
